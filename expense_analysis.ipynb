# Expense Structure Analysis – Accounting Data Product

## Analytical Problem
This notebook demonstrates the analytical workflow behind the interactive expense structure pie chart analyzer. The goal is to help accounting learners and small business managers visualize cost structures and perform "what-if" analysis on expense proportions.

**Target User**: Accounting students, cost control beginners, small business owners.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Define financial variables: expense dictionary 
expenses = {
    "labor cost": 50000,
    "material cost": 30000,
    "rent": 20000,
    "marketing costs": 15000,
    "others": 5000
}
```

```python
# Convert dictionary to pandas DataFrame 
df = pd.DataFrame({
    "Expense Category": list(expenses.keys()),
    "Amount (Yuan)": list(expenses.values())
})

# Calculate total and proportions
total = df["Amount (Yuan)"].sum()
df["Proportion (%)"] = (df["Amount (Yuan)"] / total) * 100

# Display the table
df
```

```python
# Static pie chart using matplotlib 
plt.figure(figsize=(6,6))
plt.pie(df["Amount (Yuan)"], labels=df["Expense Category"], autopct="%1.1f%%")
plt.title("Expense Structure – Static Pie Chart")
plt.show()
```

```python
# Interactive pie chart using plotly
fig = px.pie(df, values="Amount (Yuan)", names="Expense Category", title="Interactive Expense Structure")
fig.show()
```

```python
# Sort by amount descending to identify major expenses
df_sorted = df.sort_values(by="Amount (Yuan)", ascending=False)
df_sorted
```

## Key Findings
- The largest expense category is **labor cost**, accounting for the highest proportion.
- Adjusting expense amounts can significantly change the cost structure, which helps in budget planning.
- This analysis forms the basis for the interactive Streamlit tool, where users can upload their own Excel files and see real-time updates.

## Limitations
- The dataset is manually created and not from a real company.
- Only five expense categories are included; a real-world scenario may require more.
- No time-series or trend analysis is provided.
